## Capping

```python
upper_limit_height = data['Height'].mean() + 3*data['Height'].std()
lower_limit_height = data['Height'].mean()- 3 * data['Height'].std()

print('Upper limit of Height: ', upper_limit_height)
print('Lower limit of Height: ', lower_limit_height)
```

```python
upper_limit_weight = data['Weight'].mean() + 3*data['Weight'].std()
lower_limit_weight = data['Weight'].mean()- 3 * data['Weight'].std()

print('Upper limit of Weight: ', upper_limit_weight)
print('Lower limit of Weight: ', lower_limit_weight)
```

```python
new_data_height = {'new_height':np.where(data['Height']> upper_limit_height, 
                        upper_limit_height,
                        np.where(
                                    data['Height']< lower_limit_height, 
                                    lower_limit_height, 
                                    data['Height']
                                    )
                    )}
```

```python
new_data_height = pd.DataFrame(new_data_height)
```

```python
new_data_height
```

```python
new_data_weight = {'new_weight':np.where(data['Weight']> upper_limit_weight, 
                        upper_limit_weight,
                        np.where(
                                    data['Weight']< lower_limit_weight, 
                                    lower_limit_weight, 
                                    data['Weight']
                                    )
                    )}
```

```python
new_data_weight = pd.DataFrame(new_data_weight)
```

```python
new_data_weight
```

```python

fig, ax = plt.subplots(2,2, figsize = (10,8))

sns.kdeplot(data['Height'], ax=ax[0,0], color='#f5e042')
sns.kdeplot(new_data_height, ax=ax[0,1], color='#f5e042')

sns.kdeplot(data['Weight'], ax=ax[1,0], color= '#d442f5')
sns.kdeplot(new_data_weight, ax=ax[1,1], color='#d442f5')

plt.show()


```

```python

fig, ax = plt.subplots(4,2, figsize = (10,10))

sns.kdeplot(data['Height'], ax=ax[0,0], color='#d442f5')
ax[0,0].set_title("Original Height Distribution")
sns.kdeplot(new_data_height, ax=ax[0,1], color='#03fc0f')
ax[0,1].set_title("New Height Distribution")

sns.boxplot(data['Height'], ax=ax[1,0], color='#d442f5')
ax[1,0].set_title("Original Height Boxplot")
sns.boxplot(new_data_height, ax=ax[1,1], color='#03fc0f')
ax[1,1].set_title("New Height Boxplot")



sns.kdeplot(data['Weight'], ax=ax[2,0], color='#d442f5')
ax[2,0].set_title("Original Weight Distribution")
sns.kdeplot(new_data_weight, ax=ax[2,1], color='#03fc0f')
ax[2,1].set_title("New Weight Distribution")

sns.boxplot(data['Weight'], ax=ax[3,0], color='#d442f5')
ax[3,0].set_title("Original Weight Boxplot")
sns.boxplot(new_data_weight, ax=ax[3,1], color='#03fc0f')
ax[3,1].set_title("New Weight Boxplot")


fig.suptitle("Height and Weight Distribution Comparison", fontsize=16, color='#fc0f03')

plt.tight_layout()
plt.show()


```
